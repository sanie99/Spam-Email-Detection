```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pickle
```

```python
data = pd.read_csv('../data/processed/processed_spam.csv')
y = data['label_encoded']
X = data['tokens']
```

```python
# Loading the trained models
multinomial_model = pickle.load(open('../models/multinomial_model.pkl', 'rb'))
tfidf_model = pickle.load(open('../models/tfidf_model.pkl', 'rb'))
```

```python
# Vectorizing the test data
X_test_tfidf = tfidf_model.transform(X)
```

```python
# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_test_tfidf, y, test_size=0.2, random_state=42)
y_pred_train = multinomial_model.predict(X_train)
y_pred_test = multinomial_model.predict(X_test)
```

```python
# Evaluating the model on the training set
accuracy = multinomial_model.score(X_train, y_train)
print("Model Accuracy:", accuracy)

# Evaluating the model on the testing set
accuracy = multinomial_model.score(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
# Evaluating the model on the classification report
classification_rep_train = classification_report(y_train, y_pred_train)
print("Classification Report on Training Data\n", classification_rep_train)

classification_rep_test = classification_report(y_test, y_pred_test)
print("Classification Report on Test Data\n", classification_rep_test)
```

```python
# Evaluating the model on the confusion matrix
conf_matrix_train = confusion_matrix(y_train, y_pred_train)
print("Confusion Matrix on Training Data:\n", conf_matrix_train)

conf_matrix_test = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix on Test Data:\n", conf_matrix_test)
```

```python
# Evaluating the model on the accuracy score
accuracy_score_value_train = accuracy_score(y_train, y_pred_train)
print("Accuracy Score on Training Data:", accuracy_score_value_train)

accuracy_score_value_test = accuracy_score(y_test, y_pred_test)
print("Accuracy Score on Test Data:", accuracy_score_value_test)
```
